```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
#BANHEIROS
labels = 'Outros', '1 Banheiro', '2 Banheiros', '3 Banheiros'
sizes = [5.968629991209, 18.215888585573, 39.194003608939, 36.584463054643]
explode = (0, 0, 0, 0)

fig = plt.figure()
ax1 = fig.add_axes([2,2,2,2])
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  
ax1.set_title('Porcentagem de banheiros construidos nas casas')
plt.show()
print('Total: 21.613')
print('Outros: Total = 1290')
print('0 Banheiros = 0.8%,  4 Banheiros = 85.4%,  5 Banheiros = 11%,  6 Banheiros = 2%,  7 Banheiros = 0.3%,  8 Banheiros = 0.3%')
```

```python
labels = ['1901-1910', '1911-1920', '1921-1930', '1931-1940', '1941-1950', '1951-1960', '1961-1970', '1971-1980', '1981-1990', '1991-2000', '2001-2010']
construcao = [692, 770, 1184, 596, 1858, 2450, 2544, 2393, 2362, 2136, 3443]
reforma = [0, 0, 0, 3, 9, 24, 39, 58, 166, 192, 255]

x = np.arange(len(labels))  
width = 0.4

fig = plt.figure()
ax = fig.add_axes([2,2,2,2])

rects1 = ax.bar(x - width/2, construcao, width, label='Construido')
rects2 = ax.bar(x + width/2, reforma, width, label='Reformado')

ax.set_ylabel('Quantia')
ax.set_title('Casas construidas e reformadas')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 0),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
